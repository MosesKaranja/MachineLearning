```
data = df.copy()
```

```
pre_data = data.drop(columns=['SalePrice'],axis=1)
all_features = list(pre_data.columns)
len(all_features)
```

```
categorical_data = data.select_dtypes(include=['object']).copy()
categorical_features = list(categorical_data.columns)
categorical_data.head()
```

```
#data
nominal = ['MSZoning','LandContour','LotConfig','Neighborhood','RoofStyle','Functional','SaleType']
ordinal = list(set(categorical_features) - set(nominal))
numerical = list(set(all_features) - set(categorical_features))
target = ['SalePrice']

for feature in categorical_features:
  categorical_data[feature] = categorical_data[feature].astype('category')
for feature in ordinal:
  categorical_data[feature] = categorical_data[feature].cat.codes
categorical_data.head()  
   
```

```
# data_copy = categorical_data.copy()

# data_copy = pd.concat([pd.get_dummies(data_copy[nominal]),data_copy],axis=1) 
# new_data = pd.concat([data_copy,data[numerical],data[target]], axis=1)
# new_data

data_copy=categorical_data.copy()

data_copy = pd.concat([pd.get_dummies(data_copy[nominal]),data_copy],axis=1) 
data_copy.drop(nominal,axis=1,inplace=True)
#data_copy
new_data = pd.concat([data_copy,data[numerical],data[target]],axis=1)
new_data
```

```
all_features = list(new_data.columns)
all_features
X = new_data.loc[:,all_features[0]:all_features[-2]].values
y = new_data.loc[:,[all_features[-1]]].values
```

```
X = PowerTransformer().fit_transform(X)
y = PowerTransformer().fit_transform(y)
```

```
pca_with_varia = PCA(0.9)
pca_with_varia.fit_transform(X)
pca_with_varia.n_components_
```

```
pca = PCA(n_components=2)
pc_s = pca.fit_transform(X)
pca_data = pd.DataFrame(pc_s,columns=['PC1','PC2'])
target_data = pd.DataFrame(y,columns=['SalePrice'])
final_data = pd.concat([pca_data,target_data],axis=1)
final_data
```

```
pca.explained_variance_ratio_
```

```
final_data.describe()
```

```
sns.pairplot(final_data,kind='reg',diag_kind='kde')
```

```
final_data_array = final_data.to_numpy()
final_data_array
```

```
all_features = list(final_data.columns)
all_features
X_2 = final_data.loc[:,all_features[0]:all_features[-2]].values
#X_2
y_2 = final_data.loc[:,[all_features[-1]]].values
y_2
```

```
X_train, X_test,y_train,y_test = train_test_split(X_2,y_2,test_size=0.3,random_state=101)
```

```
result={}
models={'LinearRegression':linear_model.LinearRegression(),
        'Lasso':linear_model.Lasso(),
        'Ridge':linear_model.Ridge(),
        'ElasticNet':linear_model.ElasticNet()}

for i in models:
  models[i].fit(X_train,y_train)
  accuracy = models[i].score(X_test,y_test) 
  result[i] = [accuracy]
pd.DataFrame(result)          
```

```
kfold = KFold(n_splits=4)
for i in models:
  regressor = models[i]
  c_validated = cross_val_score(regressor,X_train,y_train,cv=kfold)
  accuracy = c_validated.mean()
  result[i]=[accuracy]
pd.DataFrame(result)
```

```
for i in models:
  regressor = models[i]
  if i == 'Lasso':
    lasso_params = {'alpha':[0.02,0.024,0.025,0.026,0.03],'max_iter':[1000,1100,2000],'tol':[0.001,0.0001,0.000001]}
    grid = GridSearchCV(regressor,param_grid=lasso_params,cv=2,n_jobs=-1)
    results = grid.fit(X_train,y_train)
    accuracy = results.score(X_test,y_test)
  else:
    c_validated = cross_val_score(regressor,X_train,y_train,cv=kfold)
    accuracy = c_validated.mean()
  result[i]=[accuracy]
pd.DataFrame(result)  
```

```
regressor.score(X_test,y_test)
```

```
min_val = np.min(final_data_array, axis=0)
max_val = np.max(final_data_array, axis=0)
X,Y = np.meshgrid(np.linspace(min_val[0], max_val[0], 20), np.linspace(min_val[1], max_val[1], 20))
```

```
regressor = LassoCV()
regressor.fit(X_train,y_train)
w_1 = regressor.coef_[0]
w_2 = regressor.coef_[1]
b = regressor.intercept_
Z = w_1*X + w_2*Y + b
```

```
regressor.score(X_test, y_test)
```

```
fig1 = plt.figure(figsize=(10,10))
ax = fig1.gca(projection='3d')
ax.plot_surface(X,Y,Z,rstride=1,cstride=1,alpha=0.9)
ax.scatter(final_data_array[:,0], final_data_array[:,1], final_data_array[:,2], c='r',s=50)
plt.xlabel('PC1')
plt.ylabel('PC2')
ax.set_zlabel('House Price in $')
ax.set_title("Plane of Best Fit")
```
